```python
## Build Model
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
model=smf.glm("Target~RevolvingUtilizationOfUnsecuredLines+age+C(Gender)+C(Region)+MonthlyIncome+C(Rented_OwnHouse)+C(Occupation)+C(Education)+DebtRatio+NumberOfOpenCreditLinesAndLoans+Delay_payment+NumberRealEstateLoansOrLines+NumberOfDependents",data=data,
              family=sm.families.Binomial()).fit()
```

```python
print(model.summary())
```

```python
import sklearn.metrics as metrics
```

```python
y_true = data['Target']
y_pred = model.predict(data)
```

```python
y_pred.head()
```

```python
y_pred=model.predict(data).map(lambda x:1 if x>0.5 else 0)
metrics.confusion_matrix(y_true,y_pred)
```

```python
## ROC curve
y_score=model.predict(data)
fpr,tpr,thresholds=metrics.roc_curve(y_true,y_score)
x,y=np.arange(0,1.1,0.1),np.arange(0,1.1,0.1)
```

```python
plt.plot(fpr,tpr,"-")
plt.plot(x,y,'b--')
```

```python
## AUC
metrics.roc_auc_score(y_true,y_score)
```
