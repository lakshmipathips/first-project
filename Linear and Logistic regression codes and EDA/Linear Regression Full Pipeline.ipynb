```python
data=pd.read_csv('OnlineNewsPopularity.csv')
data.describe()
```

```python
print data.columns
print data.dtypes
```

```python
data.drop('url',axis=1,inplace=True)
X=data.drop(' shares',axis=1)
y=data[' shares']
```

```python
#Predict number of shares for each article
#Split the data into test and training
import sklearn.model_selection as model_selection
```

```python
X_train,X_test,y_train, y_test = model_selection.train_test_split(X,y,test_size=0.20,random_state=400)
```

```python
#Build a linear regression model with regularization
import sklearn.linear_model as linear_model
#Lasso Model
reg=linear_model.Lasso(max_iter=1000,normalize=True)
reg=reg.fit(X_train,y_train)
```

```python
#Grid Search over Lasso model
Greg=model_selection.GridSearchCV(reg,param_grid={'alpha':np.arange(0.1,100,1).tolist()})
Greg=Greg.fit(X_train,y_train)
```

```python
Greg.best_params_
```

```python
Greg.cv_results_
```

```python
#Ridge Regression
regR=linear_model.Ridge(normalize=True)
regR=regR.fit(X_train,y_train)
```

```python
#Grid Search over Ridge model
GregR=model_selection.GridSearchCV(regR,param_grid={'alpha':np.arange(1,100,1)})
GregR=GregR.fit(X_train,y_train)
```

```python
GregR.best_params_
```

```python
#Test the models
import sklearn.metrics as metrics
import sklearn.preprocessing as preprocessing
x_test=preprocessing.normalize(X_test)
print(metrics.mean_squared_error(y_test,Greg.predict(x_test)))
print(metrics.mean_squared_error(y_test,GregR.predict(x_test)))
```
